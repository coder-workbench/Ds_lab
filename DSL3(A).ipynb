```python
import pandas as pd
data=pd.read_csv("Fortune 500 Companies.csv")
data
```

```python
data.describe()
```

```python
data['profit_mil'].describe()
```

```python
print("mean of revenue_mil is",data['revenue_mil'].mean())
print("mean of profit_mil is",data['profit_mil'].mean())
print("mean of asset_mil is",data['asset_mil'].mean())
print("mean of employees is",data['employees'].mean())
```

```python
print("mode of revenue_mil is",data['revenue_mil'].mode())
print("mode of profit_mil is",data['profit_mil'].mode())
print("mode of asset_mil is",data['asset_mil'].mode())
print("mode of employees is",data['employees'].mode())
```

```python
print("median of revenue_mil is",data['revenue_mil'].median())
print("median of profit_mil is",data['profit_mil'].median())
print("median of asset_mil is",data['asset_mil'].median())
print("median of employees is",data['employees'].median())
```

```python
print("std of revenue_mil is",data['revenue_mil'].std())
print("std of profit_mil is",data['profit_mil'].std())
print("std of asset_mil is",data['asset_mil'].std())
print("std of employees is",data['employees'].std())
```

```python
print("maximum of revenue_mil is",data['revenue_mil'].max())
print("maximum of profit_mil is",data['profit_mil'].max())
print("maximum of asset_mil is",data['asset_mil'].max())
print("maximum of employees is",data['employees'].max())
```

```python
print("minimum of revenue_mil is",data['revenue_mil'].min())
print("minimum of profit_mil is",data['profit_mil'].min())
print("minimum of asset_mil is",data['asset_mil'].min()) 
print("minimum of employees is",data['employees'].min())
```
